# Crypto Volatility Prediction - EDA

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('default')
df = pd.read_csv('../data/crypto_data.csv')
df['date'] = pd.to_datetime(df['date'])
df.head()
```

```python
df.shape, df.columns.tolist()
```

```python
df.isna().sum()
```

```python
top_cryptos = df['crypto_name'].value_counts().head(10)
top_cryptos
```

```python
plt.figure(figsize=(10,5))
sns.countplot(data=df[df['crypto_name'].isin(top_cryptos.index)], x='crypto_name', order=top_cryptos.index)
plt.xticks(rotation=45)
plt.title('Top Crypto Counts')
plt.tight_layout()
plt.show()
```

```python
sample = df[df['crypto_name'].isin(df['crypto_name'].value_counts().head(5).index)].copy()
plt.figure(figsize=(12,6))
sns.lineplot(data=sample, x='date', y='close', hue='crypto_name')
plt.title('Close Price Trend')
plt.tight_layout()
plt.show()
```

```python
eda = df[['open','high','low','close','volume','marketCap']].copy()
plt.figure(figsize=(8,6))
sns.heatmap(eda.corr(numeric_only=True), annot=True, fmt='.2f', cmap='Blues')
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.show()
```
